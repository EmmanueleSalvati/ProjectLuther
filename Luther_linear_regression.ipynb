```python
from sys import path
path.append('/Users/JerkFace/Metis/Projects/Luther/')
import pickle as pkl
import pandas as pd
import numpy as np
import dataframe_utils
import statsmodels.formula.api as sm
```

```python
with open('/Users/JerkFace/Dropbox/Metis_Dropbox/Datasets/Luther/movies_with_uncategorized.pkl', 'r') as pklfile:
    (movies, early, late) = pkl.load(pklfile)
```

### Linear regression on later movies

```python
(late_train, late_test) = dataframe_utils.train_test_samples(late)
late_train['Ones'] = 1.0
late_test['Ones'] = 1.0
```

#### Add Laplace smooting to UsersRating

```python
late_train['LaplaceRating'] = (late_train.UsersRating * late_train.NumVoters + 5.0 * 1000.) / (late_train.NumVoters + 1000)
late_test['LaplaceRating'] = (late_test.UsersRating * late_test.NumVoters + 5.0 * 1000.) / (late_test.NumVoters + 1000)
```

#### Now the regression

```python
results = pd.DataFrame(columns=['R^2', 'adjR^2', 'AIC', 'BIC', 'F', 'logL'], index=['all', 'Music', 'Drama', 'Political', 'Biography', 'History'])
```

```python
X_train = late_train[['TotalTheaters', 'LaplaceRating', 'Ones']]
Y_train = late_train['TotalGross']

late_model = sm.OLS(Y_train, X_train).fit()
# results.loc['History', :] = [late_model.rsquared, late_model.rsquared_adj, late_model.aic, late_model.bic, late_model.f_pvalue, late_model.llf]
late_model.summary()
```

## Linear Regression on early movies 

```python
(early_train, early_test) = dataframe_utils.train_test_samples(early)
early_train['Ones'] = 1.0
early_test['Ones'] = 1.0

early_train['LaplaceRating'] = (early_train.UsersRating * early_train.NumVoters + 5.0 * 1000.) / (early_train.NumVoters + 1000)
early_test['LaplaceRating'] = (early_test.UsersRating * early_test.NumVoters + 5.0 * 1000.) / (early_test.NumVoters + 1000)

X_early_train = early_train[['TotalTheaters', 'LaplaceRating',  'NumVoters', 'Ones']]
Y_early_train = early_train['TotalGross']

early_model = sm.OLS(Y_early_train, X_early_train).fit()
early_model.summary()
```

```python
plt.scatter(early_train.TotalTheaters, early_train.TotalGross)

plt.title('Training sample of movies BEFORE Farenheit 9/11\n')
plt.xlabel('Number of Theaters')
plt.ylabel('Revenue')
plt.plot(early_train.TotalTheaters, early_model.predict(), 'rs')
```

```python
X_early_test = early_test[['TotalTheaters', 'LaplaceRating',  'NumVoters', 'Ones']]
Y_early_test = early_test['TotalGross']

plt.scatter(early_test['TotalTheaters'], Y_early_test)
plt.xlim(xmin=-10)
plt.title('Test sample of movies BEFORE Farenheit 9/11')
plt.xlabel('Number of Theaters')
plt.ylabel('Revenue')
plt.plot(early_test['TotalTheaters'], early_model.predict(X_early_test), 'gs')
```

```python
plt.scatter(early_test['LaplaceRating'], Y_early_test, label='Actual')
# plt.xlim(xmin=-10)
plt.title('Test sample of movies BEFORE Farenheit 9/11')
plt.xlabel('Users\'s Rating')
plt.ylabel('Revenue')
plt.plot(early_test['LaplaceRating'], early_model.predict(X_early_test), 'gs', label='Prediction')
plt.legend()
plt.savefig('/Users/JerkFace/Metis/Projects/Luther/LR_before_RevenueVRating.pdf', frameon=True, bbox_inches='tight')
```

```python
plt.plot(Y_early_test, early_model.predict(X_early_test), 'go')
plt.plot(Y_early_test, Y_early_test, 'm--', label='Ideal Prediction')
plt.title('Over/Under Prediction of the Model BEFORE Farenheit 9/11\n')
plt.xlabel('Actual Revenue')
plt.ylabel('Predicted Revenue by Model')
plt.xlim(xmax=0.5*1e7)
plt.xticks(rotation=70)
plt.legend(loc=0)
plt.savefig('/Users/JerkFace/Metis/Projects/Luther/PredVAct_BEFORE.pdf', frameon=True, bbox_inches='tight')
plt.savefig('/Users/JerkFace/Metis/Projects/Luther/PredVAct_BEFORE.png', frameon=True, bbox_inches='tight')
```
